```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-80, 81, 0.5)
y = x**2 + 5*x + 3
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# **1-First Model**

```python
model_1 = keras.Sequential([
    keras.layers.Dense(1024, activation='relu', input_shape=[1]),
    keras.layers.Dense(1)
    ])
```

```python
model_1.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model_1.summary()
```

```python
hist_1 = model_1.fit(x_train, y_train, epochs=1000, batch_size=64, validation_data=(x_test, y_test))
```

# **2- Second Model**

```python
model_2 = keras.Sequential([
    keras.layers.Dense(512, activation='relu', input_shape=[1]),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(1)
    ])
```

```python
model_2.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model_2.summary()
```

```python
hist_2 = model_2.fit(x_train, y_train, epochs=1000, batch_size=64, validation_data=(x_test, y_test))
```

# **3- Third Model**

```python
model_3 = keras.Sequential([
    keras.layers.Dense(256, activation='relu', input_shape=[1]),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(1)
])
```

```python
model_3.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model_3.summary()
```

```python
hist_3 = model_3.fit(x_train, y_train, epochs=1000, batch_size=64, validation_data=(x_test, y_test))
```

# **Comparison**

```python
y_pred_1 = model_1.predict(x_test)
y_pred_2 = model_2.predict(x_test)
y_pred_3 = model_3.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color='black', linewidths=7)
plt.scatter(x_test, y_pred_1, color='red')
plt.scatter(x_test, y_pred_2, color='green')
plt.scatter(x_test, y_pred_3, color='blue')
plt.legend(['Exact Results', 'Prediction 1', 'Prediction 2', 'Prediction 3'], loc='upper center',
          fancybox=True, shadow=True, ncol=2) # Changed ncol to 2
plt.xlabel('X', labelpad=15)
plt.ylabel('Y', labelpad=15)
plt.title('Comparison')
plt.show()
```

```python
plt.plot(hist_1.history['loss'], label='Loss 1')
plt.plot(hist_2.history['loss'], label='Loss 2')
plt.plot(hist_3.history['loss'], label='Loss 3')
plt.legend()
```
